```python
import seaborn as sns
import matplotlib.pyplot as plt
column_names = ["selfMade", "gender", "philanthropyScore", "numberOfSiblings", "category"]
for i in column_names:
    plt.figure()
    plt.title('Count Plot of Column ' + str(i))
    sns.countplot(y = i, data = df)
    plt.figure()
    plt.title('Average Net Worth')
    if i == 'category':
        sns.barplot(y = i, x = "finalWorth", data = df)
    else:
        sns.barplot(x = i, y = "finalWorth", data = df)
```

```python
large_column_names = ["source", "country", "state", "city", "countryOfCitizenship", "organization", "title",
                     "residenceMsa"]
n = 20
for i in large_column_names:
    top_n = df[i].value_counts()[:n].index.tolist()
    top_n_df = df[df[i].isin(top_n)]
    plt.figure()
    plt.title('Count Plot for Top ' + str(n) + ' Most Common Occurences in Column ' + str(i))
    sns.countplot(y = i, data = top_n_df)
    plt.figure()
    plt.title('Net Worth')
    sns.barplot(y = i, x = "finalWorth", data = top_n_df)
```
